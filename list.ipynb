```python
# Using a tuple
a = [1, 2, 3]
b = (4, 5)
a.extend(b)
print(a) 

# Using a set
a = [1, 2, 3]
b = {4, 5}
a.extend(b)
print(a)  

# Using a string
a = ['a', 'b']
b = "cd"
a.extend(b)
print(a)
```

List Comprehension in Python

```python
a = [2,3,4,5]
res = [val ** 2 for val in a]
print(res)
```

```python
a = [2,3,4,5]
res = [val ** 2 for val in a]
print(res)
```

```python
a = [1, 2, 3, 4, 5]
res = [val * 2 for val in a]
print(res)
```

```python
a = [1, 2, 3, 4, 5]
res = [val for val in a if val % 2 == 0]
print(res)
```
