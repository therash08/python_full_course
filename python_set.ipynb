```python
s1 = {4, 12, 10, 9, 4, 13}

print("Minimum element: ", min(s1))      
print("Maximum element: ", max(s1))
```

```python
s = {5, 3, 9, 1, 7}

# Sorting the set (converts it into a sorted list)
sorted_s = sorted(s)

print("Minimum element: ", sorted_s[0])
print("Maximum element: ", sorted_s[-1])
```

```python
s = {5, 3, 9, 1, 7}

# Initialize min and max with extreme values
min_val = float('inf')
max_val = float('-inf')

# Iterate through the set to find min and max
for ele in s:
    if ele < min_val:
        min_val = ele
    if ele > max_val:
        max_val = ele

print("Minimum element:", min_val)
print("Maximum element:", max_val)
```

```python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

res1 = a | b
print("Union:", res1)


res2 = a & b
print("Intersection:", res2)

res2 = a.union(b)
print("Union using method:", res2)

res3 = a.intersection(b)
print("Intersection using method:", res3)

res4 = a - b
print("Difference (a - b):", res4)

```
