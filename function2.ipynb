1. positional arguments
2. keyword arguments
3. default arguments
4. variable arguments


positional arguments

```python
def add_num(x,y):
    result =  x+y
    return result

add_num(10,30)
    
```

keyword arguments

```python
def personal(job, income):
    print(job)
    print(income)
personal('data science', 2000)
```

default arguments

```python
def info(name, age):
    print("Name:", name)
    print("Age:", age)

info(age=22, name="Rafi")
```

```python
def greet(name="Guest"):
    print("Hello,", name)

greet("Rafi")   # আর্গুমেন্ট পাঠানো হয়েছে
greet()         # পাঠানো হয়নি, তাই "Guest" ব্যবহার হবে
```

variable length arguments

```python
def add_num(*x):
    print(x)
    result = sum(x)
    return result
    
add_num(10,20,1000,2000)
```

```python
def personal_data(NAME, **arg):
    print(NAME)
    print(arg)
personal_data('rasdul hoque', uni =  'lu', city = 'syl')
```
